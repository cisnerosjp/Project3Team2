```python
top_ten_countries = df.sort_values(by= ['AQI Value', 'Country'], ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 8))
plt.bar(top_ten_countries['Country'], top_ten_countries['AQI Value'], color='blue')
plt.xlabel('Country')
plt.ylabel('AQI Value')
plt.title('Top Ten Countries by AQI Value')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

```python
sns.set(rc={'figure.figsize': (16, 9)})
g = df.groupby('Country', as_index=False)['AQI Value'].mean().sort_values(by='AQI Value', ascending=False).head(10)
ax = sns.barplot(data=g, x='Country', y='AQI Value', hue='AQI Value', dodge=False)
ax.set(yticklabels=[])
for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)
plt.xlabel('Country')
plt.ylabel('AQI Value')
plt.title('Top Ten Countries by AQI Value')
plt.show()
```

```python
sns.set(rc={'figure.figsize': (16, 9)})
g = df.groupby('Country', as_index=False)['AQI Value'].mean().sort_values(by='AQI Value', ascending=True).head(10)
ax = sns.barplot(data=g, x='Country', y='AQI Value', hue='AQI Value', dodge=False)
ax.set(yticklabels=[])
for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10)
plt.xlabel('Country')
plt.ylabel('AQI Value')
plt.title('Top Ten Countries by AQI Value')
plt.show()
```
